LINEAR REGRESSION

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
```

```python
df = pd.read_csv('data1.csv') 

df_binary = df[['Pretest_Result','Test_Result','SSC_Result']]

df_binary.columns = ['Pretest_Result','Test_Result','SSC_Result']

df_binary.head()
```

```python
sns.lmplot(x ="Pretest_Result", y ="SSC_Result", data=df_binary,ci=None)
```

```python
sns.lmplot(x="Test_Result", y="SSC_Result", data=df_binary,ci=None) 
```

```python
y=df_binary['SSC_Result']

x = zip(df_binary['Pretest_Result'],df_binary['Test_Result'])

X = []
for i in x:
    X.append(i)
    
    
    
X=np.array(X)
y=np.array(y)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

regr = LinearRegression()

regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))

#print(y)
#print(X)
```

```python
t1 = eval(input("Enter Pre_Test Result :"))
t2 = eval(input("Enter Test Result :"))
p=[[t1,t2]]
```

```python
SSC = regr.predict(p)
for i in SSC: 
    #print(i, end="")
    print("{0:.2f}".format(i),'%')
```
